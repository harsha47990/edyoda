```python
rawdata.drop(['Year','OG','Dept','Union'],axis = 1,inplace=True)
rawdata
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()
rawdata['JF'] = le.fit_transform(rawdata['JF'])
rawdata['DC'] = le.fit_transform(rawdata['DC'])
rawdata['Job'] = le.fit_transform(rawdata['Job'])
rawdata['YT'] = le.fit_transform(rawdata['YT'])
```

```python
X_data = rawdata.iloc[:,0:10]
Y_data = rawdata['Total_Compensation']
```

```python
X_data
```

```python
Y_data
```

```python
ss = StandardScaler()
X_data = ss.fit_transform(X_data)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_data,Y_data,test_size = 0.2,random_state=12)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rms = []
for i in range(1,25):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train,Y_train)
    pred = knn.predict(X_test)
    error = sqrt(mean_squared_error(Y_test,pred))
    rms.append(error)
    
```

```python
sns.lineplot(x=range(1,25),y = rms)
```
