```python
groups = {}
for cat in rawdata['StratificationCategory1'].unique():
    temp = rawdata[rawdata['StratificationCategory1']==cat]
    groups[cat] = temp.groupby('YearStart')['Data_Value'].agg('mean')
```

```python
#predictions based on gender over years
#predicting % of male and female having  obesity by 2035
genderwise = rawdata[(rawdata['Class'] =='Obesity / Weight Status') & (rawdata['StratificationCategory1']=='Gender')]
```

```python
genderwiseGroup = genderwise.groupby(['YearStart','Stratification1'])['Data_Value'].agg(mean='mean')
genderwiseGroup.reset_index(inplace=True)
```

```python
genderwiseGroup
```

```python
import seaborn as sns
sns.catplot(data=genderwiseGroup,x = 'YearStart',y= 'mean' ,hue = 'Stratification1')
```

```python
import numpy as np
X_data = genderwiseGroup.copy()
X_data["Stratification1"] = np.where(X_data["Stratification1"] == "Female", 0, 1)
```

```python
X_data
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_data.iloc[:,0:2],X_data.iloc[:,2])
```

```python
lr.predict([[2035,1],[2035,0]])
```

```python
#predictions based on age range over years
Agewise = rawdata[(rawdata['Class'] =='Obesity / Weight Status') & (rawdata['StratificationCategory1']=='Age (years)')]
```

```python
Agewise = Agewise.groupby(['YearStart','Age(years)'])['Data_Value'].agg(mean='mean')
Agewise.reset_index(inplace=True)
Agewise
```

```python
from sklearn.preprocessing import LabelEncoder
le_agewise = LabelEncoder()
Agewise['Age(years)'] = le_agewise.fit_transform(Agewise['Age(years)'])
```

```python
Agewise
```

```python
lr = LinearRegression()
lr.fit(Agewise.iloc[:,0:2],Agewise.iloc[:,2])
```

```python
lr.predict([[2035,1]])
```
